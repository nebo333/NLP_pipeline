```python
from ipynb.fs.full.Preproc_pipeline_Text import *

from ipynb.fs.full.Algo_Kmeans_w_textnormer_and_vectorizer import *
```

```python
pic_root = '/Users/nebo333/Documents/Pickled files/'
cat_pattern = r'reca[\w]+.*' ## limiting to just recautos.
pic_corpus = TextPickledCorpusReader(pic_root, cat_pattern)
```

```python
docs = pic_corpus.docs()
```

```python
from sklearn.pipeline import Pipeline

model = Pipeline([
    ('norm', TextNormalizer()),
    ('vect', OneHotVectorizer()),
    ('clusters', KMeansClusters(k=7))
])

clusters = model.fit_transform(docs)
pickles = list(pic_corpus.fileids())
for idx, cluster in enumerate(clusters):
    print("Document '{}' assigned to cluster {}.".format(pickles[idx],cluster))
```
